```python
# # Ticket_price where origin is not equal to Destination 

def nrst_to_mean(li):
    me=np.mean(li)
    nrs=min(range(len(li)),key=lambda x:abs(li[x]-me))
    return li[nrs]

AADT={i:{m:{q:nrst_to_mean(df[(df["Airline"]==i)&
                          (df["Aircraft_Type"]==m) &
                          (df["Distance_km"]==q) &
                          (df["Ticket_Price"].notna())]
                          ["Ticket_Price"].tolist()) for q in n 
            }
            for m,n in {k:df[(df["Airline"]==i) &
                             (df["Aircraft_Type"]==k) &
                             (df["Distance_km"]!=0.0)]
                             ["Distance_km"].unique().tolist() for k in j
                        }.items()
        }
            for i,j in{i:df[(df["Airline"]==i)]["Aircraft_Type"].unique()
                for i in df['Airline'].unique()
                        }.items()
    }

for i,j in AADT.items():
    for m,n in j.items():
        for q,w in n.items():
            df.loc[(df["Airline"]==i) &
                   (df["Aircraft_Type"]==m) &
                   (df["Distance_km"]==q) &
                   (df["Ticket_Price"].isna()),"Ticket_Price"]=w
            
# Ticket price where origin is equal to destination
OD={o:{d:df[(df["Origin"]==o) & 
            (df["Destination"]==d)]
            ["Distance_km"].unique().tolist()[0] for d in ld
        }for o,ld in 
        {i: df[(df["Origin"]==i) & 
               (df["Origin"]!=df["Destination"])]
               ["Destination"].unique().tolist()
            for i in df["Origin"].unique()
        }.items()
    }

 
for i,j in AADT.items(): 
    for k,l in j.items():
        # print(i,k,m,n) 
         for o in OD:
            des=np.random.choice(list(OD[o].keys()))
            dis=OD[o][des] 
            # print(i,k,o,des,dis, AADT[1][k][dis])
            df.loc[(df["Airline"]==i) & 
                   (df["Aircraft_Type"]==k) & 
                   (df["Origin"]==o) & 
                   (df["Origin"]==df["Destination"]) & 
                   (df["Ticket_Price"].isna()),"Ticket_Price"]=AADT[i][k][dis]
```

```python

AADWF={i:{j:{d:df[(df["Airline"]==i) &
            (df["Aircraft_Type"]==j) & 
            (df["Distance_km"]==d) & 
            (df["Fuel_Cost"].notna())].groupby("Weather_Condition")["Fuel_Cost"].unique().to_dict() 
            for d in df[(df["Airline"]==i) & 
                        (df["Aircraft_Type"]==j)].groupby("Distance_km")["Weather_Condition"].unique().to_dict()
            }              
                        for j in df[df["Airline"]==i].groupby("Aircraft_Type")["Distance_km"].unique().to_dict()
        }
                        for i in df.groupby("Airline")["Aircraft_Type"].unique().to_dict()
    }


for i,j in AADWF.items(): 
    for k,l in j.items(): 
        for m,n in l.items(): 
            for q,w in n.items(): 
                #print(i,k,m,q, np.random.choice(w))  
                df.loc[(df["Airline"]==i) & 
                       (df["Aircraft_Type"]==k) & 
                       (df["Distance_km"]==m) & 
                       (df["Weather_Condition"]==q) &                       
                       (df["Fuel_Cost"].isna()),"Fuel_Cost"]=np.random.choice(w)
```

```python
# DURATION hours

AADWS={i:{j:{d:{w: (df[(df["Airline"]==i) &
                        (df["Aircraft_Type"]==j) &
                        (df["Distance_km"]==d) &
                        (df["Weather_Condition"]==w) &
                        (df["Duration_hours"].notna())]["Distance_km"] / df[(df["Airline"]==i) & 
                                                                            (df["Aircraft_Type"]==j) &
                                                                            (df["Distance_km"]==d) &
                                                                            (df["Weather_Condition"]==w) &
                                                                            (df["Duration_hours"].notna())]["Duration_hours"]).to_list()
                for w in q
            }for d,q in df[(df["Airline"]==i) &
                            (df["Aircraft_Type"]==j) &
                            (df["Distance_km"]!=0.0)].groupby("Distance_km")["Weather_Condition"].unique().to_dict().items()
        }for j in df[(df["Airline"]==i)].groupby("Aircraft_Type")["Distance_km"].unique().to_dict() 
    }for i in df.groupby("Airline")["Aircraft_Type"].unique().to_dict()
}
```

```python
for i,j in AADWS.items():
    for l,k in j.items():
        for q,w in k.items():
            for p,o in w.items():
                try:
                    df.loc[(df["Airline"]==i) &
                           (df["Aircraft_Type"]==l) &
                           (df["Distance_km"]==q) &
                           (df["Weather_Condition"]==p) &
                           (df["Duration_hours"].isna()),"Duration_hours"]=q/np.random.choice(o)
                except Exception as e:
                    pass

df.loc[(df["Distance_km"]==0.0) & 
       (df["Duration_hours"].isna()),"Duration_hours"]=df[(df["Distance_km"]==0.0) & 
                                                         (df["Duration_hours"].notna())]["Duration_hours"].mean()
```

```python
def dhf(a):
    h=int(a)
    m=int((a-h)*60)
    s=round(((a-h)*60-m)*60)
    if s>=60:
        m+=s//60
        s=s%60
    if m>=60:
        h+=m//60
        m=m%60
    return f"{h:02d}:{m:02d}:{s:02d}"
df["Duration_hours"]=df["Duration_hours"].apply(lambda x:dhf(x))
```

```python
def dmf(a):
    h=0
    m=int(a)
    s=round((a-m)*60)
    if s>=60:
        m+=s//60
        s=s%60
    if m>=60:
        h+=m//60
        m=m%60
    return f"{h:02d}:{m:02d}:{s:02d}"
df["Delay_minutes"]=df["Delay_minutes"].apply(lambda x:dmf(x))
```
